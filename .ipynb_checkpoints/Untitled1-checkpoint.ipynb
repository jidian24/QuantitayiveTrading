```python
# vectorbt 简书示例保留
import pdb
import vectorbt as vbt
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
ohlc_df = pd.read_csv('..//data//Gate.io_ETC_swap_1d.csv',index_col=0,parse_dates=True)  # 表格列名：datetime,open,hight,low,volume
price =ohlc_df.close
# 计算移动平均线
# 使用 vectorbt 库的 MA 函数对价格数据进行 10 期快速移动平均线计算
fast_ma = vbt.MA.run(price, 7) 
# 对价格数据进行 50 期慢速移动平均线计算
slow_ma = vbt.MA.run(price, 26) 
# 获取快速移动平均线向上穿越慢速移动平均线的信号
entries = fast_ma.ma_crossed_above(slow_ma) 
# 获取快速移动平均线向下穿越慢速移动平均线的信号
exits = fast_ma.ma_crossed_below(slow_ma) 
pf_kwargs = dict(size=np.inf, fees=0.001, freq='1D')  # 定义投资组合的参数字典：本金，费用，交易频率
pf = vbt.Portfolio.from_signals(price, entries, exits, direction='both', **pf_kwargs)  # 根据参数创建投资组合对象

print(pf.trades.records_readable)  # 打印每笔交易盈亏明细
print(pf.orders.records_readable)    # 打印交易信号
print(entries.value_counts())    # 打印入场信号数量
# pf.trades.plot().show_svg()
print(pf.stats())
# pf.trades.plot().show()    # 绘制交易信号图
# 绘制交易信号图和快慢双均线


    
fig = pf.trades.plot()    # 绘制交易信号图

# 添加快慢均线到图中
fig.add_scatter(x=price.index, y=fast_ma.ma, mode='lines', name='Fast MA (7)')
fig.add_scatter(x=price.index, y=slow_ma.ma, mode='lines', name='Slow MA (26)')
fig.show()    # 显示图表


pdb.set_trace()
fig = pf.orders.plot(width=1000)  #绘制下单信号图示
fig.show()

# 绘制 heatmap 图 , y_level='slow_window' slider_level='symbol', 
fig = pf.total_return().vbt.heatmap(
    x_level='fast_window', symmetric=True,
    trace_kwargs=dict(colorbar=dict(title='Total return', tickformat='%')))

fig.show()
```

```python
import pdb	
import pandas as pd  
import pandas_ta as ta  
import wind_library
import vectorbt as vbt

if __name__ == '__main__':
    df = pd.read_csv('..//data//Gate.io_ACA_swap_1d.csv')  # 表格列名：datetime,open,hight,low,volume
    sma_short= 7; sma_long= 26
    momentumStrategy = ta.Strategy(
        name="Custommomentum",
        description="roc,obv,",
        ta=[
            {"kind": "roc", "close": "volume", "prefix": "vol"}, #"length": 10 ,
            {"kind": "obv"},
            {"kind": "rsi"} ,
            {"kind": "kdj"} ,
            {"kind": "sma", "length": sma_short},
            {"kind": "sma", "length": sma_long},
            {"kind": "bbands"} ,
            {"kind": "macd", } 
        ]
    )
    signalStrategy = ta.Strategy(
        name="signalStrategy",
        description="signals",
        ta=[
            #{"kind": "roc", "close": "volume", "prefix": "vol"}, #"length": 10 ,
            #{"kind": "obv"},
            #{"kind": "rsi"} ,
            #{"kind": "sma", "length": 7},
           {"kind": "cross", "close": "sma_" + str(sma_short), "signal": "sma_" + str(sma_long)} ,
            {"kind": "above", "close": "close", "signal": "sma_" + str(sma_short)}
        ]
    )
    df.ta.strategy(momentumStrategy)
    df.ta.strategy(signalStrategy)
    pdb.set_trace()
    # 调用动态绘图函数
    wind_library.finance_plot(df, ['datetime', 'open', 'high', 'low'])
```

```python
# from PyPDF2 import PdfFileReader, PdfFileWriter
import tkinter as tk
from tkinter import filedialog

def pdf_txt_replace(pdf_files, text1, text2):
    for pdf_file in pdf_files:
        with open(pdf_file, "rb") as file:
            pdf_reader = PdfFileReader(file)
            pdf_writer = PdfFileWriter()

            for page_num in range(pdf_reader.getNumPages()):
                page = pdf_reader.getPage(page_num)
                page_text = page.extract_text()

                # 替换文本
                new_page_text = page_text.replace(text1, text2)
                page.merge_page(page)
                page.merge_page(page)
                pdf_writer.addPage(page)

            # 保存替换后的 PDF 文件
            output_file = pdf_file.replace(".pdf", "_replaced.pdf")
            with open(output_file, "wb") as output:
                pdf_writer.write(output)

            print(f"已将文件 {pdf_file} 中的 '{text1}' 替换为 '{text2}'，并保存为 {output_file}")

def select_files():
    root = tk.Tk()
    root.withdraw()
    file_path = filedialog.askopenfilenames(title="选择PDF文件", filetypes=[("PDF files", "*.pdf")])
    return file_path

# 通过对话框选择待处理的 PDF 文件
pdf_files = select_files()

# 替换文本
text1 = "三垦力达电气（江阴）有限公司"
text2 = ""
pdf_txt_replace(pdf_files, text1, text2)
```
