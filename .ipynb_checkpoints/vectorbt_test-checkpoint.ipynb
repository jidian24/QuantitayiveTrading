```python
# vectorbt example
import pdb
import ccxt
import numpy as np
import pandas as pd
import wind_library
import vectorbt as vbt


# Gate.ioBTC/USDT1K
btc_price = wind_library.get_ccxt_exchange_kline('gate', 'BTC/USDT', 'spot', '1d', 100).close
etc_price = wind_library.get_ccxt_exchange_kline('gate', 'ETC/USDT', 'spot', '1d', 100).close
comb_price = btc_price.vbt.concat(etc_price, keys=pd.Index(['BTC', 'ETC'], name='symbol'))
comb_price.vbt.drop_levels(-1, inplace=True)
# Multiple strategy instances: (10, 30) and (20, 30)
fast_ma = vbt.MA.run(comb_price, [10, 20], short_name='fast')
slow_ma = vbt.MA.run(comb_price, [30, 30], short_name='slow')
# 
entries = fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_below(slow_ma)
pf = vbt.Portfolio.from_signals(comb_price, entries, exits, direction='both', size=np.inf, fees=0.001, freq='1h', sl_stop=0.01, tp_stop=0.02)
print(f'total_return:\n{pf.total_return()}')    # 
mean_return = pf.total_return().groupby('symbol').mean()     #  'symbol' 
mean_return.vbt.barplot(xaxis_title='Symbol', yaxis_title='Mean total return')     # x  'Symbol'y  'Mean total return'
# print(price_etc)
```

```python
# vectorbt example
import pdb
import ccxt
import numpy as np
import pandas as pd
import wind_library
import vectorbt as vbt


# Gate.ioBTC/USDT1K
btc_price = wind_library.get_ccxt_exchange_kline('gate', 'BTC/USDT', 'spot', '1d', 100).close
etc_price = wind_library.get_ccxt_exchange_kline('gate', 'ETC/USDT', 'spot', '1d', 100).close
comb_price = btc_price.vbt.concat(etc_price, keys=pd.Index(['BTC', 'ETC'], name='symbol'))
comb_price.vbt.drop_levels(-1, inplace=True)
# Multiple strategy instances, instruments, and time periods
mult_comb_price, _ = comb_price.vbt.range_split(n=2)   # 
mult_comb_price
# pdb.set_trace()
# Multiple strategy instances: (10, 30) and (20, 30)
fast_ma = vbt.MA.run(mult_comb_price, [10, 20], short_name='fast')
slow_ma = vbt.MA.run(mult_comb_price, [30, 30], short_name='slow')
# 
entries = fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_below(slow_ma)
pf = vbt.Portfolio.from_signals(mult_comb_price, entries, exits, direction='both', size=np.inf, fees=0.001, freq='1d', sl_stop=0.01, tp_stop=0.02)
pf.total_return()
# pdb.set_trace()
print(f'total_return:\n{pf.total_return()}')    # 
mean_return = pf.total_return().groupby(['split_idx', 'symbol']).mean()     #  'symbol' 
mean_return.unstack(level=-1).vbt.barplot(xaxis_title='Split index', yaxis_title='Mean total return', legend_title_text='Symbol')     # x  'Symbol'y  'Mean total return'
# mean_return = pf.total_return().groupby(['split_idx', 'symbol']).mean()     #  'symbol' 
# mean_return.unstack(level=-1).vbt.barplot(xaxis_title='Split index', yaxis_title='Mean total return', legend_title_text='Symbol')
```

```python
import os
import pdb	
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt

exchange = getattr(ccxt, 'gate')()
# vectorbt
df = pd.read_csv('..//data//Gate.io_ACA_swap_1d.csv')  # datetime,open,hight,low,volume

# Create the "Golden Cross" 	
df["GC"] = df.ta.sma(50, append=True) > df.ta.sma(200, append=True)
pdb.set_trace()
# Create boolean Signals(TS_Entries, TS_Exits) for vectorbt	vectorbt(TS_Entries, TS_Exits)
golden = df.ta.tsignals(df.GC, asbool=True, append=True)

# Sanity Check (Ensure data exists)		
print(df)

# Create the Signals Portfolio	
pf = vbt.Portfolio.from_signals(df.close, entries=golden.TS_Entries, exits=golden.TS_Exits, freq="D", init_cash=100_000, fees=0.0025, slippage=0.0025)	#freq=

# Print Portfolio Stats and Return Stats	
print(pf.stats())
print(pf.returns_stats())
df.to_csv(os.path.join('..', 'data', 'Return.csv'), index=True)
```

```python
# -*- coding: utf-8 -*-
import io
import sys

    # 
def save_package_help(package_name):
    package = __import__(package_name)
    all_functions = [func for func in dir(package) if callable(getattr(package, func))]

    #  io.StringIO  help 
    help_output = io.StringIO()

    #  help_output
    sys.stdout = help_output

    # 
    help_text = ""
    for func_name in all_functions:
        try:
            help_text += f"# {func_name}\n\n"  # 
            help(getattr(package, func_name))
            help_text += help_output.getvalue() + "\n\n"  # 
            help_output.truncate(0)  #  help_output
            help_output.seek(0)
        except:
            pass

    # 
    sys.stdout = sys.__stdout__

    #  .txt 
    file_path = f"{package_name}_help.txt"
    with io.open(file_path, "w", encoding="utf-8") as file:
        file.write(help_text)

    print(f" {file_path}")

# example 
# save_package_help("vectorbt")
```
