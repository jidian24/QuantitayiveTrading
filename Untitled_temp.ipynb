```python
import os
import pdb	
import pandas as pd
import pandas_ta as ta
import vectorbt as vbt
# vectorbt回测示例
df = pd.read_csv('..//data//Gate.io_ACA_swap_1d.csv')  # 表格列名：datetime,open,hight,low,volume

# Create the "Golden Cross" 	创建“黄金交叉”。
df["GC"] = df.ta.sma(50, append=True) > df.ta.sma(200, append=True)
pdb.set_trace()
# Create boolean Signals(TS_Entries, TS_Exits) for vectorbt	为vectorbt创建布尔信号(TS_Entries, TS_Exits)
golden = df.ta.tsignals(df.GC, asbool=True, append=True)

# Sanity Check (Ensure data exists)		合理性检查（确保数据存在）
print(df)

# Create the Signals Portfolio	创建信号组合
pf = vbt.Portfolio.from_signals(df.close, entries=golden.TS_Entries, exits=golden.TS_Exits, freq="D", init_cash=100_000, fees=0.0025, slippage=0.0025)	#freq=频率

# Print Portfolio Stats and Return Stats	打印投资组合统计数据和回报统计数据
print(pf.stats())
print(pf.returns_stats())
df.to_csv(os.path.join('..', 'data', 'Return.csv'), index=True)
```

```python
import mplfinance as mpf
import pandas as pd

# 创建示例数据
data = pd.read_csv('..//data//Gate.io_ACA_swap_1d.csv', index_col=0, parse_dates=True)
data.index.name = 'datetime'

# 计算移动平均线
data['MA7'] = data['close'].rolling(window=7).mean()
data['MA26'] = data['close'].rolling(window=26).mean()

# 创建蜡烛图并添加技术指标
add_plot = [
    mpf.make_addplot(data['MA7'], color='b', title='MA7'),
    mpf.make_addplot(data['MA26'], color='r', title='MA26')
]

mpf.plot(data, type='line', style='yahoo', title='Candlestick Chart with Moving Averages', ylabel='Price', addplot=add_plot)
```
